```python
#verb_lst = []
all_past_stmt = []
temp_pst = ""

for test_case in list(df["Expect_res"]):
    stmt_lst = custom_sent_tokenizer.tokenize(test_case)
    try:
        for stmt in stmt_lst: 
            words = nltk.word_tokenize(stmt) 
            verb_lst = [word for (word, pos) in nltk.pos_tag(words) if is_verb(pos)] 
            temp_pst = temp_pst + '\n' + transform_stmt(stmt, verb_lst)
        all_past_stmt.append(temp_pst)
        temp_pst = ""

    except Exception as e:
        all_past_stmt.append(test_case)
        print(str(e))
```
